# Timestamp를 이용한 시간별 분석

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/track1.csv')
df.Timestamp = pd.to_datetime(df.Timestamp)
```

## Timestamp 컬럼을 Day, Hour, Minute, Second로 나눕니다.

```python
df['Day'] = df['Timestamp'].dt.day
df['Hour'] = df['Timestamp'].dt.hour
df['Minute'] = df['Timestamp'].dt.minute
df['Second'] = df['Timestamp'].dt.second
```

```python
df.head()
```

### 총 5일간에 로그 기록이 존재 (24, 25, 26, 27, 28)
- 기록된 로그는 24일이 37318건으로 가장 적고 27일이 471272건으로 가장 많습니다.

```python
x = df['Day'].value_counts().index
y = df['Day'].value_counts()
plt.title('Days from log')
plt.xlabel('Days')
plt.bar(x, y)
plt.show()
print(y)
```

### 일별 시간(Hours) 로그 분포
- 24일 : 0 ~ 9시 접근기록이 없습니다.
- 25일, 26일, 27일 : 1 ~ 9시 접근기록이 없습니다.
- 28일 : 0 ~ 9시, 20 ~ 24시 접근기록이 없습니다.
- 24일 20시가 1089건으로 가장 적고 27일 14시가 99708건으로 가장 많은 접근이 시도되었습니다.

```python
days = df['Day'].unique()
plt.figure(figsize=(16,9))
result = {}
for i in range(5):
    x = df[df['Day'] == days[i] ]['Hour'].value_counts().index
    y = df[df['Day'] == days[i] ]['Hour'].value_counts()
    result[days[i]] = y
    plt.subplot(3, 3, i+1)
    plt.title(f'Hours from {days[i]}')
    plt.xlabel('Hours')
    plt.ylim([0,100000])
    plt.tight_layout()
    plt.bar(x, y)
plt.show()
```
